```python
# configuration
vector_db_path = r'E:\python\Jupyter\vector_db' 
collection_name = 'doc_collection'
```

```python
# loading the embedding model - default model
embedding = HuggingFaceEmbeddings()
```

```python
# initialize groq llm
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.0
)
```

```python
vector_store = Chroma(
    collection_name=collection_name,
    embedding_function=embedding,
    persist_directory=vector_db_path
)
```

```python
retriever = vector_store.as_retriever()
```

```python
test_vector = embedding.embed_query("Is this model working?")
print(f"Vector length: {len(test_vector)}")
print(f"Sample values: {test_vector[:5]}")
```

```python
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True
)
```

```python
query = "What is Energy Flow?"
response = qa_chain.invoke({"query":query})
print(response["result"],'\n')
print("*-*-"*20,'\n')
for source in response["source_documents"]:
    print(source.metadata)
```
